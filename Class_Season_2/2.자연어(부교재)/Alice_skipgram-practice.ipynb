```python
from keras.models import Model
from keras.layers import Input,Dense,Dropout
import numpy as np
import keras.backend as K
from textUtils import textPreprocessing
import collections
import nltk
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
txt=open(r'./alice_in_wonderland.txt')
lines=[]
for line in txt:
    if len(line) <= 1:
        continue
    lines.append(textPreprocessing(line))
txt.close()
```

```python
lines #Vocabulary 만들자 이제
```

```python
lst=[]
for i,j in enumerate(lines):
    for a in j.split():
        lst.append(a)
print(lst)
```

```python
voca=set(lst)
print(voca) #vocabulary?
```

```python
#dictionary 만들기
dic={}
for i, j in enumerate (voca):
    dic[j]=i+1
print(dic)
```

```python
idx2word = {v:k for k,v in dic.items()}
```

```python
#가장 많이 쓰인 애들 순으로 일련번호 붙여보자 ------>나중에 해보자.
counter = collections.Counter()
for line in lines:
    for word in nltk.word_tokenize(line):
        counter[word.lower()] += 1
print(counter)
```

```python
#학습 데이터 만들자
xs=[]
ys=[]

for line in lines:
    embed = [dic[w.lower()] for w in nltk.word_tokenize(line)]
    triples = list(nltk.trigrams(embed))
    w_letf=[x[0] for x in triples]
    w_center=[x[1] for x in triples]
    w_right=[x[2] for x in triples]
    xs.extend(w_center)
    ys.extend(w_letf)
    xs.extend(w_center)
    ys.extend(w_right)
    
print(len(dic))
voca_size=len(dic)+1 #왜 +1?
```

```python
onehot=OneHotEncoder(categories=[range(voca_size)])
```

```python
X=onehot.fit_transform(np.array(xs).reshape(-1, 1)).todense()#todense(행렬을 sum)
Y=onehot.fit_transform(np.array(ys).reshape(-1, 1)).todense()
```

```python
Xtrain,Xtest,Ytrain,Ytest,xstr,xsts=train_test_split(X,Y,xs,test_size=0.3, random_state=42)
print(Xtrain.shape, Xtest.shape, Ytrain.shape, Ytest.shape)
```

```python
np.random.seed(42)

Batch_size=128
epoch=20
```

```python
K.clear_session()

x_input=Input(shape=(Xtrain.shape[1],))
h1=Dense(300, activation='relu')(x_input)
drop1=Dropout(0.5)(h1)
h2=Dense(2, activation='relu')(drop1)
drop2=Dropout(0.5)(h2)
h3=Dense(300, activation='relu')(drop2)
drop3=Dropout(0.5)(h3)
x_output=Dense(Ytrain.shape[1], activation='softmax')(drop3)

model=Model(x_input, x_output)

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(Xtrain,Ytrain,batch_size=Batch_size,epochs=epoch, verbose=1, validation_split = 0.2)
```

```python
#word embedding vector 값 추출
encoder=Model(x_input,x_output)
reduce_x=encoder.predict(Xtest)
```

```python
#시각화
pf = pd.DataFrame(reduce_x)
pf.columns = ['xaxis','yaxis']
pf['word_index'] = xsts
pf['word'] = pf['word_index'].map(idx2word)
pf.head()
```
