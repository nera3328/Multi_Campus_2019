```python
import nltk
```

```python
text=nltk.corpus.gutenberg.raw('austen-emma.txt')
print(text[:500])
print('문자 개수 = ' , len(text))
```

```python
emma=nltk.corpus.gutenberg.words('austen-emma.txt')
print(emma)
print('word 개수 = ', len(emma))
```

```python
sentence=nltk.corpus.gutenberg.sents('austen-emma.txt')
for i in range(5):
    print(sentence[i])
print('문장 개수 = ', len(sentence))
```

```python
longestlen=max(len(s) for s in sentence)
logestsent=[s for s in sentence if len(s)==longestlen]
print(logestsent)
```

```python
import matplotlib.pyplot as plt
import numpy as np
countword=[len(n) for n in sentence]
n, bin, patches = plt.hist(countword, bins=50)
plt.show()
print('문장 당 평균 단어 수 = %.2f' %np.mean(countword))
```

```python
from urllib import request
url='http://WWW.gutenberg.org/files/1342/1342-0.txt'
response=request.urlopen(url)
text=response.read().decode('utf8')
print(text[:500])
```

```python
tokens=nltk.word_tokenize(text)
print('토큰 타입 = ', type(tokens))
print(tokens[:10])

nltktext=nltk.Text(tokens)
print('문서 타입 = ', type(nltktext))
print(nltktext[:10])
```
