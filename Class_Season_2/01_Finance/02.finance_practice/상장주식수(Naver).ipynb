#### 아래 사이트에서 시가 총액 순으로 상장주식수를 읽어온다.
https://finance.naver.com/sise/sise_market_sum.nhn

```python
import pandas as pd
from urllib.request import urlopen
import bs4
```

```python
url = 'https://finance.naver.com/sise/sise_market_sum.nhn'
source = urlopen(url).read()
```

```python
source = bs4.BeautifulSoup(source, 'lxml')
```

```python
tbody = source.find_all('tbody')
```

```python
tbody[0].find_all('tr')[1].find_all('td')[7].text
```

#### 5라인 마다 라인이 그어져 있으므로 6번째 부터는 이를 고려해 주어야 한다.

```python
n = 1
d = 6
i = 1
siga = pd.DataFrame(columns=['code','name','amount'])
for i in range(50):
    stock_code = tbody[0].find_all('tr')[n].find_all('td')[1].find_all('a')[0]['href'].split('=')[1]
    stock_name = tbody[0].find_all('tr')[n].find_all('td')[1].text
    stock_amnt = tbody[0].find_all('tr')[n].find_all('td')[7].text.replace(',','')
    n += 1
    if n == d:
        n += 3
        d += 8
        
    # 우선주는 제외한다.
    if stock_code[-1] == '0':
        siga.loc[i] = ['a'+ stock_code, stock_name, stock_amnt]
```
