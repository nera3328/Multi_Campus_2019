```python
import pandas as pd
import numpy as np
import tensorflow as tf

saveDir='./Saver/predictStock(reg).save'
saveFile=saveDir+'/save'
```

```python
KT=pd.read_csv('./stockData/030200.csv',index_col=0, parse_dates=True)
                                     # 0번째 열 인덱스화  index 날짜를 파싱할까?
KT=KT.sort_index()
KT=KT.dropna()
KT=KT/10000
```

```python
KT['kta120'] = KT['close'].rolling(window=20).mean()      # window=120: 120일 이동평균  rolloing: 데이터 아래로 sliding    mean: 평균
KT['kta240'] = KT['close'].rolling(window=60).mean()      # 240일 이동평균
```

```python
KT['last close']=KT['close'].shift(-1)
KT.tail()
```

```python
KT['class']= 0.0
KT.loc[KT['last close']> KT['close'], 'class']=1
KT=KT.drop('volume',axis=1)
```

```python
KT=KT.sample(frac=1)
KT
```

```python
testX = np.array(KT.iloc[-1:, 0:6])

KT=KT.dropna()

#X_Input=np.array(KT.iloc[:, 0:4])
#X_Input.shape

#Y_output=np.array(KT.loc[:, 'class']).reshape(-1,1)
#Y_output

inputX = np.array(KT.iloc[:1500, 0:6])
outputY = np.array(KT.iloc[:1500, 7]).reshape(-1,1)

testX = np.array(KT.iloc[1500:, 0:6])
testY = np.array(KT.iloc[1500:, 7]).reshape(-1,1)
```

```python
outputY
```

```python
InPut=6
We=10
OutPut=1
```

```python
tf.reset_default_graph()

x = tf.placeholder(tf.float32, shape=[None,InPut],name='x')
y = tf.placeholder(tf.float32, shape=[None,OutPut], name='y')

W1 = tf.Variable(tf.truncated_normal([InPut, We]), dtype=tf.float32, name= 'W1')
B1 = tf.Variable(tf.zeros([We]),dtype=tf.float32, name= 'B1')
H1 = tf.nn.relu(tf.add(tf.matmul(x,W1), B1, name= 'H1'))

#W2 = tf.Variable(tf.truncated_normal([We,We1]), dtype=tf.float32, name= 'W2')
#B2 = tf.Variable(tf.zeros(We1),dtype=tf.float32, name= 'B2')
#H2 = tf.nn.sigmoid(tf.matmul(H1, W2) +B2, name= 'H2')
    
W0 = tf.Variable(tf.truncated_normal([We, OutPut]), dtype=tf.float32, name='W0')
B0 = tf.Variable(tf.zeros([OutPut]),dtype=tf.float32, name='B0')

lstY = tf.nn.sigmoid(tf.add(tf.matmul(H1,W0),B0, name='lstY'))

cost = tf.reduce_mean(tf.square(tf.subtract(lstY, y)))

learning_rate = 1
alpha = tf.placeholder(tf.float32, shape=(), name='alpha')
optimizer = tf.train.GradientDescentOptimizer(alpha)
train = optimizer.minimize(cost)


#saver=tf.train.Saver()
```

```python
nBatchCnt= 5
nBatchSize= int(inputX.shape[0] / nBatchCnt)

sess = tf.Session()
sess.run(tf.global_variables_initializer()) #그래프 실행

#if tf.train.checkpoint_exists(saveDir):
#    saver.restore(sess, saveFile)
```

```python
for i in range(10000):
    # Mini-batch 방식으로 학습한다
    for n in range(nBatchCnt):
        # input 데이터를 Mini-batch 크기에 맞게 자른다
        nFrom = n * nBatchSize
        nTo = n * nBatchSize + nBatchSize

        # 마지막 루프이면 nTo는 input 데이터의 끝까지.
        if n == nBatchCnt - 1:
            nTo = inputX.shape[0]
            
        bx = inputX[nFrom : nTo]
        by = outputY[nFrom : nTo]
        
        _, xcost = sess.run([train, cost], feed_dict={x: bx, y: by, alpha:0.001})
    
    if i % 1000 == 0:
        print('cost = %.6f' % xcost)
           
# 학습 결과를 저장해 둔다
#saver.save(sess, saveFile)
```

```python
#yHat[np.where(yHat>0.5)]=1
#yhat[np.where(yHat<0.5)]=0
yHat=sess.run(lstY, feed_dict={x: testX})
print(yHat)
yHat=np.array([1 if x > 0.5 else 0 for x in yHat]).reshape(-1,1)
```

```python
sm = pd.read_csv('./stockData/005930.csv', index_col=0, parse_dates=True)
sm = sm.sort_index() # 날짜 순으로 정렬한다
sm = sm.dropna()
sm=sm/10000 #스케일을 줄여서 
```

```python
sm['next close'] = sm['close'].shift(-1)
```

```python
sm.tail() #옆에 새로 생김!
```

```python
testX = np.array(sm.iloc[-1:, 0:4]) # 맨 아래 줄 애들 가리킴, 행과 열로 원하는 애들 뽑아옴
testX
```

```python
sm = sm.dropna() #마지막 날 nan값 없앰
```

```python
inputX = np.array(sm.iloc[:,0:4])# input 데이터-> (1999, 4) 구조로 바꿈
outputY =np.array(sm.loc[:, 'next close']).reshape(-1,1)  # 1차원 구조로 맞춰주기 위해서 reshape해줌
                            # 이름으로 바로 지정
```

```python
outputY #2차원 구조 이면서도 하나씩 정렬됨.-------------->inputX와 outY 정리 끝!
```

```python
nInput = 4      # input layer의 neuron 개수
nHidden = 80     # hidden layer의 neuron 개수
nOutput = 1     # output layer의 neuron 개수
```

```python
# 그래프를 생성한다
tf.reset_default_graph()

x = tf.placeholder(tf.float32, shape=[None, nInput], name='x')
y = tf.placeholder(tf.float32, shape=[None, nOutput], name='y')
```

```python
#Hidden 1
# hidden layer의 Weight (Wh)와 Bias (Bh)
W1 = tf.Variable(tf.truncated_normal([nInput, nHidden]), dtype=tf.float32, name='W1')
B1 = tf.Variable(tf.zeros(nHidden), dtype=tf.float32, name='B1')

#Hidden 2
#hidden_layer가 많으면 sigmoid 말고 relu로! sigmoid는 1과 0으로 구분하기 때문에 가격값을 찾는 것에 적합하지 않다.
W2 = tf.Variable(tf.truncated_normal([nHidden, nHidden]), dtype=tf.float32, name='W2')
B2 = tf.Variable(tf.zeros(nHidden), dtype=tf.float32, name='B2')

# output layer의 Weight (Wo)와 Bias (Bo)
Wo = tf.Variable(tf.truncated_normal([nHidden, nOutput]), dtype=tf.float32, name='Wo')
Bo = tf.Variable(tf.zeros(nOutput), dtype=tf.float32, name='Bo')

H1 = tf.nn.relu(tf.matmul(x, W1) + B1, name='H1') 
H2 = tf.nn.relu(tf.matmul(H1, W2) + B2, name='H2')


# Output layer의 출력값. activation function은 sigmoid
predY = tf.nn.sigmoid(tf.add(tf.matmul(H2, Wo), Bo, name='predY'))

# Cost function 정의. cross-entropy 사용
cost = tf.reduce_mean(tf.square(tf.subtract(predY, y)))

# 학습
optimizer = tf.train.AdamOptimizer(0.05)
train = optimizer.minimize(cost)

alpha=tf.placeholder(tf.float32, shape=(), name='alpha')
```

```python
# 그래프를 실행한다
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10000):
    _, xcost = sess.run([train, cost], feed_dict={x: inputX, y: outputY, alpha:0.01})
    
    if i % 1000 == 0:
        print('cost = %.4f' % xcost)
```

```python
yHat = (sess.run(predY, feed_dict={x: testX}))*10000
print('\ny=추정치 = \n', yHat)
```

```python
print("\ny-추정치 = \n", yHat)

# Weight, Bias의 최종값을 확인해 본다
rW1, rB1, rWo, rBo = sess.run([W1, B1, Wo, Bo])

print("\nW1 = \n", rW1)
print("\nB1 = \n", rB1)
print("\nWo = \n", rWo)
print("\nBo = \n", rBo)
sess.close()
```
