```python
webtoon_url = 'https://comic.naver.com/webtoon/weekdayList.nhn'


weekday = ['mon','tue','wed','thu','fri','sat','sun']

title = []
cartoonist = []
grade = []


for week in weekday: 
    param = {'week':week}
    result = requests.get(webtoon_url,param)
    soup = BeautifulSoup(result.content,'lxml')
    tag_ul = soup.find_all('ul', class_='img_list')
    tag_li = tag_ul[0].find_all('li')
    
        
    for li in tag_li:
        tag_dl = li.select("dl")
        tag_dt_a = tag_dl[0].select("dt a")   #title
        tag_dd_a = tag_dl[0].select("dd a")   #작가
        tag_strong = tag_dl[0].select("strong")  #평점
        
       
        
        for i in range(len(tag_dt_a)):
            title.append(tag_dt_a[i]['title'])
            cartoonist.append(tag_dd_a[i].text)
            grade.append(tag_strong[i].text)
#             print(tag_dt_a[i]['title'])
#             print(tag_dd_a[i].text)
#             print(tag_strong[i].text)
```
