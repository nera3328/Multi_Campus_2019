#### 색 및 스타일 지정

- 색 : red, blue, green, yellow = r,b,g,y,    검정/하양 = k,w, 청록/자홍 = c,m
- 선 스타일 : 실선 = -, dashed line = --, 점선 = :, dash-dot = -.
- 선 두께   : linewidth = 숫자로 조정
- marker(점) : 원 = o(알파벳), 세모 = ^/v/</>, 네모 = s(square), 별 = *, 십자 = +, 다이아 = D(알파벳), x = x
- marker(점) 크기: s = 숫자로 조정

```python
plt.plot([0,1],[0,10], c = 'y')  # c -> color, r,b,y,g
plt.show()
```

- 그래프 안에 격자를 넣고 싶다면?

```python
year = [i for i in range(2010,2020)]
price = [i for i in range(1,11)]
plt.scatter(year,price, c='k', marker = '^', s = 50)
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc = 0)
plt.grid()
plt.show
```

- 하나의 그래프 안에 여러 개의 plot을 넣고 싶다면?

```python
year = [i for i in range(2010,2020)]
price_1 = [i for i in range(1,11)]
price_2 = [i for i in range(11,1,-1)]

plt.scatter(year,price_1, c='k', marker = '^', s = 50, alpha = 0.5)  # alpha는 불투명도
plt.plot(year,price_2, c='k', linestyle = ':', linewidth = 1, marker = 'o')
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc = 0)
plt.grid()
plt.show
```

### Bar Plot

```python
member = ['ardino', 'minivelo','aladin']
before = [30,50,10]
after = [50,60,11]
```

```python
index = len(member)
index = np.arange(len(member))  #arange(시작 = 0, 끝+1 = 3, 간격=1)
index
```

```python
plt.bar(index,before, tick_label = member, color = ['r','g','b'], alpha = 0.5, width = 0.3)
plt.show()
```

```python
plt.barh(index,after, tick_label = member, color = ['r','g','b'], alpha = 0.5)   #수평바
plt.show()
```

### 히스토그램

```python
import random

for i in range(10) :
    number = random.randint(1,10)
    print(number, end=' ')
```

```python
#1-100 사이의 score 100개를 만들어보자
score = [random.randint(1,100) for num in range(100)]     #num을 안쓴다 -> 플레이스 홀더를 이용해서 표현 가능
```

```python
score = [random.randint(1,100) for _ in range(100)]  #플레이스 홀더
```

```python
plt.hist(score)  #bin=10
plt.show
```

```python
plt.hist(score,bins=20 )  #binary를 20으로 정해주자  -> 분포 안에서 20등분
plt.show
```

#### pie(파이)

```python
label = ['a','b','c','d']
percent = [40, 30, 20, 10]
plt.pie(percent, labels = label,autopct = '%.1f%%')  # autopct = '%.1f%%' -> 소수점 첫째자리까지 뽑자
plt.savefig('graph.png')  # 그래프 저장
plt.show
```

- 그래프의 크기를 조절하고 싶다면

```python
label = ['a','b','c','d']
percent = [40, 30, 20, 10]
plt.figure(figsize = (5,5))  #그래프 사이즈
plt.pie(percent, labels = label,autopct = '%.1f%%')  # autopct = '%.1f%%' -> 소수점 첫째자리까지 뽑자
plt.savefig('graph.png')  # 그래프 저장
plt.show
```
