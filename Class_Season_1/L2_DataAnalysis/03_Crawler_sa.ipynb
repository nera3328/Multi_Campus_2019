### 웹사이트에 접속해보기

- URL이란?  
네트워크상에서 자원(웹페이지)을 요청하는 규약.

- URL구조  
프로토콜://주소(orIP):포트번호/리소스경로?쿼리스트링   
ex) http://, 포트번호 -> 해당 주소로 갈건데 어느 문으로 갈래?에서 문,  
근데 프로토콜 = 포트번호 임 http = 80이다.

- 프로토콜  
인터넷에서 주로 사용하는 프로토콜은 HTTP이다. FTP,SFTP,SSH등이 있다.
- 주소 or IP  
데이터를 요청하는 대상.
- 포트번호  
프로토콜마다 각자의 고유 포트번호가 있다. HTTP는 80번이다.
- 리소스경로  
해당 사이트에서의 페이지 값.  
-> 그 사이트가 하나의 페이지로 이루어진것이 아닌 여러페이지가 모인것 그중 메인 페이지가 홈페이지  
-> 주소다음에/(슬래시)로시작하는부분.
- 쿼리스트링  
서버에 보내는 데이터값. 리소스경로 다음에 ?(물음표)로 시작하는 부분.  
변수1=데이터1&변수2=데이터2…이런 형식으로 되어있다.

###  BeautifulSoup

- requests를 이용해서 읽어들인 웹 데이터를 파싱(Parsing)하고 싶다면?  
request로 읽어들인 정보를 request.txt로 저장해서, str을 다루는 메서드로 할 수 있다.   그러나.....  
너무 불편, 이를 편하게 해주는 오픈 소스가 BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
search_url = 'http://search.naver.com/search.naver'
keyword = '정상회담'
param = {'where':'news', 'query':keyword}   #퀘리스트링에 where라는 키가 있는 것을 직접 찾아보고 사용

result = requests.get(search_url,param)
result.status_code
```

https://search.naver.com/search.naver?where=news&query=
여기서 ? 뒤에 있는게 parameter
parameter는 그 사이트에 접속할 수 있는 지표
parameter를 알 수 있는 방법은 그 사이트에 들어가서 url을 직접 보거나 그 사이트를 만드는 사람들에게 물어보는것

request = 패키지(요청하기 위한)  
get은 request 중 하나인 함수(요청하기 위한 방법 중 하나)

```python
Soup = BeautifulSoup(result.content, 'lxml')  #BeautifulSoup(대상, '파서')
```

```python
print(result.content)   # 파싱 안한거
print(Soup)             # 파싱 한거
# 이 둘을 비교해봄으로서 파싱을 하고 안하고의 차이를 알 수 있다.
```

.contents 의 타입은 리스트  ->  html 소스 코드 전체가 하나의 리스트
.text 의 타입은 str

```python
now_ul = []
count = 0
for u in ul_contents.contents:  #.contents 해준 데이터의 타입은 태그 -> 이거 처리해주려면 str으로 바꿔줘야한다.
    if not str(u).strip():  # str(u)가 strip이 되면 append, 안되면 그냥 continue(pass), 빈칸이면 false여서 continue(그냥 패스), 
        continue
    now_ul.append(u)        # 빈칸이 아니라 문자열이 존재하면 위에 만든 빈리스트에 문자열 u를 더해준다.
    count += 1
print(now_ul, count)        # ul_contents.contents 안에 몇개의 원소가 들었는지 궁금해서 count 한듯

#<li 어쩌구저쩌구> </li>  -> 하나의 리스트?
# 중간중간에 공백 존재하는데 페이지에 그럼 ul태그로 싸여진 공백이 존재해서 그런거
```

-리스트랑 딕셔너리로 만드는 것도 해보자

```python
Music_url='https://music.naver.com/listen/top100.nhn?domain=TOTAL_V2'#URL설정
Music=requests.get(Music_url)                                        #GET방식으로 REQUSTS요청한것을 Music에 저장
Soap=BeautifulSoup(Music.text,'lxml')                                #저장된 MUSIC을 text로 변환하여 'lxml'파서를 이용해 BeautifulSoup에담아 Soup에 저장
title= Soap.select('a._title span.ellipsis')                         #저장된 Soup에 a태그에 class값 _title,span태그에 class값 ellipsis를 찾아 title에 저장
music_list=[]                                                        #빈 list생성
for s in title:                                                      #title에 저장된것의 수만큼 반복되는 s는 title에 저장된 반복횟수번쩨 값
    music_dict={}                                                    #빈 사전 생성
    music_dict['name']=s.text                                        #사전에 name이라는 컬럼을 생성해서 s의 text를 넣겠다
    music_list.append(music_dict)                                    #리스트에 위에저장한 사전의 값을 붙여넣겠다
print(music_list)                                                    #리스트 출력
```
