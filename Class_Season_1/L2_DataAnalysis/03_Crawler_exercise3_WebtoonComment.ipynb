링크를 가져왔는데 /로 시작하면 그 링크는 리소스경로 로 프로토콜과 주소값 필요
- href 는 a 태그의 하이퍼링크 url이 들어가 있다.
- 그런데 href 값을 보니 /부터 시작
- /      = 리소스 경로 (?     = 쿼리스트링)  -> 리소스경로 뒤에 쿼리스트링 붙은거

**결론**
- '프로토콜://주소값/' 가져와서 이거 뒤에 href값(리소스경로) 붙이자.
- 어떤 프로토콜: //주소값을 긁어와야되는지는 어떻게 알지? = href 클릭해보면 안다.

```python
url = 'https://comic.naver.com/webtoon/weekday.nhn'

week = ['mon','tue','wed','thu','fri','sat','sun']
request = requests.get(url)

re_cont = request.content

re_parsing = BeautifulSoup(re_cont,'lxml')

title = []
id_list = []
tag_day = re_parsing.select('div.col_inner')

for d in tag_day:
        tag_week = d.select('h4')
        for w in tag_week:
            print(w.text)
        
        tag_title = d.select('a.title')
        for t in tag_title:
            print(t.text)
            a = t['href']
            print(a)
            title_id = re.findall('titleId=(.+?)&weekday', a)
            id_list.append(title_id)
            print(title_id)
            
        
        
        print()
```
