### 색 및 스타일 지정

- 색 : red/blue/green/yellow/black/white/청록/자홍 = r/b/g/y/k/w/c/m
- 스타일 : 실선 = -, dashed line = --, 점선 = :, dash-dot line = -.
- 선 두께 : linewidth= 숫자로 조정, 
- 마커 : 원 = o, 세모 = ^/v/</>, 네모 = s, 별 = *, 십자 = + ,다이아 = D, x = x
- 마커(점) 크기 : s= 숫자로 조정
- 불투명도 alpha=숫자(0.0~1)

```python
plt.plot([0, 1,2], [0,10,20], '-')
plt.plot([2, 1, 0], [0,10,20], c='r', marker='D')
plt.show()
```

- legend(범례) : 0 (best, 최적위치 선정). 1(우측상단),2(좌측상단), 3(좌측하단), 4(우측하단)


```python
size = [30,40,50]
price = [1.5,2.4,4.6]
plt.scatter(size,price, c='g')
plt.xlabel('size')
plt.ylabel('price')
plt.title('price-size')
plt.legend('data_1', loc = 4)
plt.show()
```

- 그래프 안에 격자를 넣고 싶다면? plt.grid()

```python
year=[i for i in range(2010, 2020)]
price=[i for i in range(1,11)]
plt.scatter(year,price, c='k', marker='D')
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc=0)
plt.grid()
plt.show()
```

- 하나의 그래프 안에 여러개의 plot을 넣고 싶다면?

```python
year=[i for i in range(2010, 2020)]
price_1=[i for i in range(1,11)]
price_2=[i for i in range(11,1, -1)]


plt.scatter(year,price_1, c='k', marker='D', s=50, alpha= 0.1)
plt.plot(year,price_2, c='r', linestyle=':', linewidth=7, marker='^')
plt.xlabel('year')
plt.ylabel('price')
plt.title('price-year')
plt.legend('y', loc=0)
plt.grid()
plt.show()
```

```python
city= ['seoul', 'daejeon', 'daegu', 'busan']
#위도(lat), 경도(lon)
lat =[37,36,35,35]
lon =[127,128,129,130]
#인구밀도
pop_den = [16154, 2839, 2790, 4454]
```

quiz) 위도와 경도에 맞추어서 도시를 scatter 이용하여 점으로 표현하되, 인구밀도에 비례하여 크기를 다르게 표현해보세요

```python

plt.scatter(lon, lat, c=['r','b','k','m'], alpha=0.5,marker='o', s=pop_den)
plt.xlabel('위도')
plt.ylabel('경도')
plt.title('인구밀도')
plt.show()
```

```python
for x,y,d in zip(lon,lat, pop_den) :
    plt.scatter(x,y, s=d)
```

- city 이름도 함께 표현하려면?(TEXT)

```python
plt.scatter(lon, lat, s=pop_den)
for x,y,c in zip(lon, lat, city) : 
    plt.text(x,y,c)
plt.show()
```

```python
for x,y,d,c in zip(lon,lat, pop_den,city) :
    plt.scatter(x,y, s=d, alpha=0.4)
    plt.text(x,y,c, alpha=0.6)
plt.show()
```

### Bar Plot


```python
member= ['ardino','minivelo','aladin']
before = [30, 50, 10]
after = [50, 60, 11]
```

```python
index = np.arange(len(member)) #arange(시작 = 0, 끝+1=3, 간격=1)
```

```python
plt.bar(index,before, tick_label=member,color=['r','b','g'], alpha=0.5, width=0.3)
plt.show()
```

```python
plt.barh(index, after, tick_label=member, color=['r','b','g'],alpha=0.5, height=0.3)
plt.show()
```
