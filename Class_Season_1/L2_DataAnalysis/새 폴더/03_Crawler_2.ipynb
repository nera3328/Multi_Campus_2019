```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.songok.es.kr/segio/meal_v2/meal_month.php'

params = {'year': 2019, 'month':5}
result = requests.get(url, params)

soup = BeautifulSoup(result.content, 'lxml')

tag_div = soup.find('div', class_='food_mon')
#     print(tag_div)
tag_tr = tag_div.find_all('tr')

meals = []
for tr in tag_tr:
    tag_td = tr.find_all('td')
    for td in tag_td:
        tag_a = td.select('a')
        if len(tag_a) != 0:
            meals.append(tag_a[0].text)

temp = []
for meal in meals: 
    t = re.findall('[가-힣]+', meal)
#     print(t)
    if t:
        start = re.findall('.+밥', t[0])
        if start:
            temp.append(t)

for t in temp:
    print(t)

# new_meals = []
# for meal in meals: #emal = 하루의 식단 
#     temp = meal.replace('\n', ' ').split()
#     new_meal = []
#     for t in temp:
#         if not '.' in t and not '[' in t:
#             new_meal.append(t)
#     if new_meal:
#         new_meals.append(new_meal)

# import pprint
# pprint.pprint(new_meals, indent=4)
```

#### 기사제목(정규표현식)

```python
search_url = "https://search.naver.com/search.naver"
params= {'query' : "이하민"}


result = requests.get(search_url, params=params)

html = result.text
html
```

```python
#format , end 파라미터 복습 -> 04_Input & Output 앞부분

tag_dl = re.findall('<dl> <dt>(<a.+?</a>)</dt> <dd', html, re.DOTALL)

print('tag_a의 개수 : {}개'.format(len(tag_dl)), end='\n\n---------------\n\n')

for i, dl in enumerate(tag_dl):
    print('{}번째 a 태그\n'.format(i))
    print(dl, end='\n\n')
```

```python
# 03_If&Loop 복습

link = []
for dl in tag_dl:
    link.append(re.findall('<a href="(.+?)" target', dl,
                           re.DOTALL))

import pprint as pp    
pp.pprint(link)
```

#### bs4과 re(정규표현식)

beautifulsoup을 기본으로 하면서,  
찾을 태그나 속성을 정규표현식으로 표현하여 사용할 수 있다.  
  
  - re.compile('정규표현식') : 패턴 지정
  - re.findall(re.compile('정규표현식'), 찾을 위치(doc))

```python
doc = 'abcabc123'

pattern = re.compile('[0-9]+')
data = re.findall(pattern, doc)
print(data)

pattern = re.compile('[0-9]+')
data_2 = pattern.findall(doc)
print(data_2)

data_3 = re.findall('[0-9]+', doc) #compile로 지정하는 작업 생략
print(data_3)
```

```python
search_url = "https://search.naver.com/search.naver"
params= {'where': 'news', 'query' : "이하민"}


result = requests.get(search_url, params=params)

Soup = BeautifulSoup(result.content, 'lxml')

#정규표현식
tag_ul = Soup.find('ul', class_=re.compile('t.+?1')) # = type01

tag_li = []
for con in tag_ul.contents:
    if not str(con).strip():
        continue
    tag_li.append(con)

for li in tag_li:
    a_tag = li.find('dl').find('dt').find('a')

    print(a_tag['href'])
    print(a_tag.text)
```
