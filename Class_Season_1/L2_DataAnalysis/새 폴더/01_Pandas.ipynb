```python
df_country = pd.DataFrame({'name' : ['알라딘', '자스민', '자카'], '국적':['아라비아', '아라비아', '아라비아']})
df_country
```

```python
data = df.merge(df_country, how = 'right') # how = left, right, inner, outer
data

#inner : 서로 같은 index만 뽑아서 merge하자. (how값의 default는 inner)
#outer : 전부 다 합치고 겹치는 값은 merge, 없는 값은 NaN으로 merge
#left : df.merge(df_country) / 왼쪽인 df에서 row(행)을 일단 다 쓰고 여기에 right를 합치자.
#rigth : 위와 반대로
```

```python
df_left = pd.DataFrame({'key' : ['A', 'B', 'C'], 'left' : [1,2,3]})
df_left
```

```python
df_right = pd.DataFrame({'key' : ['A', 'B', 'D'], 'right' : [4,5,6]})
df_right
```

```python
test = df_left.merge(df_right, how='left')
test
```

```python
test = df_left.merge(df_right, how='right')
test
```

```python
test = df_left.merge(df_right, how='outer')
test
```

- merge(on) #기준열 설정

```python
test = df_left.merge(df_right, how='right', on = '')
test
```

- left, right 모두에게 'data'라는 이름의 columns를 추가하되, left와 right에 다른 값을 넣어보세요

```python
df_left2 = pd.DataFrame({'key' : ['A', 'B', 'C'], 'left' : [1,2,3], 'data' : ['100','200','300']})
df_right2 = pd.DataFrame({'key' : ['A', 'B', 'D'], 'right' : [4,5,6],'data' : ['a','b','c']})
```

```python
df = df_left2.merge(df_right2, on='key')
df
```

- 반대로 기준열로 삼아야하는 컬럼이 서로 이름이 다른경우

```python
df_left3 = pd.DataFrame({'key' : ['A', 'B', 'C'], 'left' : [1,2,3]})
df_right3 = pd.DataFrame({'name' : ['A', 'B', 'D'], 'right' : [4,5,6]})
```

```python
df = df_left3.merge(df_right3, left_on='key', right_on='name', how ='outer')
df
```

```python
df = pd.concat([df_left3, df_right3]) #concat은 값을 아래로 쌓은다.
df
```

```python
df = pd.concat([df_left3, df_right3], axis=1) #concat은 값을 아래로 쌓은다, '0'이 default. axis=1은 옆으로쌓인다.
df
```
