# Convert datetime and add some useful columns

```python
pd.to_datetime(data['Date/Time']).head()
```

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
# pd.to_datetime : 해당 컬럼을 datetime 형태로 변환시키기
# map function : bulit-in 함수, list/dictionary 같이 iterable한 데이터를 인자로 받아 list 안의 개별 
#item을 함수의 인자로 전달해서 결과를 list형태로 반환함
```

```python
data.head()
```

```python
# 함수만들기

# def function_name(function_args):
#    functions
#    return output_args


# 함수 예제
def add_two_int(x,y):
    value = x+y
    return value

add_two_int(1,2)
```

```python
def get_day(dt):
    return dt.day

data['date'] = data['Date/Time'].map(get_day)
```

```python
data
```

```python
def get_wkday(dt):
    return dt.weekday()

def get_hour(dt):
    return dt.hour

data['weekday'] = data['Date/Time'].map(get_wkday)
data['hour']    = data['Date/Time'].map(get_hour)

data.tail()
```

# Analysis

```python
import matplotlib.pyplot as plt
```

```python
# histogram of day of month
plt.hist(data.date, bins=30, rwidth=.8, range=(.5,30.5))
plt.xlabel('date of the month')
plt.ylabel('frequency')
plt.title('Frequency by day of month - uber - Apr 2014')
```

```python
def count_rows(rows):
    return len(rows)

by_date = data.groupby('date').apply(count_rows)
by_date
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
plt.bar(range(1,31), by_date_sorted)
plt.xticks(range(1,31), by_date_sorted.index)
plt.xlabel('date of month')
plt.ylabel('frequency')
plt.title('Frequency by day of month - uber - Apr 2014')
```

# difference btw map, apply, applymap

pandas에서 배열의 합이나 평균 등 일반적인 통계는 DataFrame내 함수를 사용하면 되지만, pandas에서 제공하지 않는 기능, 즉 내가 만든 커스텀 함수(custom function)를 DataFrame에 적용하려면 map, apply, applymap 함수를 사용함.

1. map 함수는 DataFrame 타입이 아니라, 반드시 Series 타입에서만 사용해야 함
  시리즈 클래스(Series) = 값(value) + 인덱스(index)
  - series는 값 하나하나에 인덱스를 붙인 데이터
  - series의 값을 하나씩 꺼내서 lambda 함수의 인자로 넘기는 커스텀 함수를 각 값 별로 실행
  
2. apply 함수는 커스텀 함수를 사용하기 위해 DataFrame에서 복수 개의 컬럼이 필요할 때 apply함수를 사용

3. applymap 함수
  - apply함수처럼 각 row나 각 column별로 작동하는 함수가 아니라, 각 요소별로 작동함
