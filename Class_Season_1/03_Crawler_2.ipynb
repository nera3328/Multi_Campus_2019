#### 기사제목(정규표현식)

```python
search_url = "https://search.naver.com/search.naver"
params= {'query' : "이하민"}


result = requests.get(search_url, params=params)

html = result.text
html
```

```python
#format , end 파라미터 복습 -> 04_Input & Output 앞부분

tag_dl = re.findall('<dl> <dt>(<a.+?</a>)</dt> <dd', html, re.DOTALL)

print('tag_a의 개수 : {}개'.format(len(tag_dl)), end='\n\n---------------\n\n')

for i, dl in enumerate(tag_dl):
    print('{}번째 a 태그\n'.format(i))
    print(dl, end='\n\n')
```

```python
# 03_If&Loop 복습

link = []
for dl in tag_dl:
    link.append(re.findall('<a href="(.+?)" target', dl,
                           re.DOTALL))

import pprint as pp    
pp.pprint(link)
```

#### bs4과 re(정규표현식)

찾을 태그나 속성을 정규표현식으로 표현하여 사용할 수 있다.  
  
  - re.compile('정규표현식') : 패턴 지정
  - re.findall(re.compile('정규표현식'), 찾을 위치(doc))

```python
doc = 'abcabc123'

pattern = re.compile('[0-9]+')
data = re.findall(pattern, doc)
print(data)

pattern = re.compile('[0-9]+')
data_2 = pattern.findall(doc)
print(data_2)

data_3 = re.findall('[0-9]+', doc) #compile로 지정하는 작업 생략
print(data_3)
```

```python
search_url = "https://search.naver.com/search.naver"
params= {'where': 'news', 'query' : "이하민"}


result = requests.get(search_url, params=params)

Soup = BeautifulSoup(result.content, 'lxml')

#정규표현식
tag_ul = Soup.find('ul', class_=re.compile('t.+?1')) # = type01

tag_li = []
for con in tag_ul.contents:
    if not str(con).strip():
        continue
    tag_li.append(con)

for li in tag_li:
    a_tag = li.find('dl').find('dt').find('a')

    print(a_tag['href'])
    print(a_tag.text)
```

- BeautifulSoup의 find와 같이 객체 1개 반환

1) 루트 태그와 루트가 아닌 태그
- 루트 태그 : /태그명 ex) /div -> 루트 div
- 루트가 아닌 태그 : //태그명 ex) //div -> 모든 div
  
2) //태그명[@속성명=속성값] ex) //div[@id='abc']

3) (//태그명)[3] -> 3번째 '태그명'태그 ex) (//a)[3] -> 3번째 a
- (//태그명)[last()] -> 마지막 태그

4) 아스테리크 = \*(전부)   
ex) //div[@\*] = 어떤 속성이라도 가지고 있는 div  
ex) //table/tr/\* = 모든 tr

naver 검색해주는 함수를 한번 만들어보세요.

```python
def naver_search(search_word): #검색어를 입력인자로 넣어주세요.
    driver = webdriver.Chrome(driver_path)
    url = 'https://www.naver.com'
    driver.get(url)
    
    select_input = driver.find_element_by_xpath('//*[@id="query"]')   
    select_input.clear()
    select_input.send_keys(search_word)
    select_input.send_keys(Keys.ENTER)
    
    #연관검색어
    relative_words = driver.find_elements_by_xpath('//*[@id="nx_related_keywords"]/dl/dd[1]/ul/li/a')
    results = []
    for i in relative_words:
        if i.text:
            results.append(i.text)
    return results
```

```python
input_word = '아이유'
relative_words = naver_search(input_word)
print(relative_words)
```
