```python
search_url = 'http://search.naver.com/search.naver'
keyword = '정상회담'

param = {'where': 'news', 'query': keyword}

result = requests.get(search_url, param)
result.status_code
```

```python
ul_contents = Soup.find('ul', class_='type01')
len(ul_contents.contents)
```

```python
news_list = []
count = 0
for li in now_ul:
    news_dict = {}
    a_tag = li.find('dl').find('dt').find('a')
    
    if count == 0:
        print(a_tag, end='\n\n')
    count += 1
    news_dict['link']=a_tag['href']
    
    detail_resp = requests.get(news_dict['link'])
    detail_resp = BeautifulSoup(detail_resp.content, 'lxml')
    
    news_dict['title'] = a_tag.text
#     print(a_tag.text)
        
#     print(news_dict, end='\n\n')
    news_list.append(news_dict)
# print(len(news_list), news_list)
# for i, news in enumerate(news_list):
#     print(i, news)
```

```python
search_url = "https://search.naver.com/search.naver"
query = "레드벨벳"

for i in range(10):
    params = {
        'where': 'news',
        'query': query,
        'start': (i*10) + 1
    }

    result = requests.get(search_url, params=params)

    soup = BeautifulSoup(result.content, 'lxml')
    
    tag_ul = soup.find('ul', class_='type01')
    
    tag_li = []
    for con in tag_ul.contents:
        if not str(con).strip():
            continue
        tag_li.append(con)
    
    for li in tag_li:
        a_tag = li.find('dl').find('dt').find('a')
    
        print(a_tag['href'])
        print(a_tag.text)
```

```python
search_url = "https://comic.naver.com/webtoon/weekdayList.nhn"

weekday = ['mon','tue','wed','thu','fri','sat','sun']

title, author, score, image = [], [], [], []
for week in weekday:
    params = {'week': week}

    result = requests.get(search_url, params)
    soup = BeautifulSoup(result.content, 'lxml')
    tag_ul = soup.find_all('ul', class_="img_list")
    tag_li = tag_ul[0].find_all('li')
    for li in tag_li:
        tag_dl = li.select("dl")
        tag_dt_a = tag_dl[0].select("dt a") #호랑이형님
        tag_dd_a = tag_dl[0].select("dd a") #이상규
        tag_strong = tag_dl[0].select("strong") #9.98
        tag_img = li.select("div.thumb a img")
        
        title.append(tag_dt_a[0].text)
        author.append(tag_dd_a[0].text)
        score.append(tag_strong[0].text) 
        image.append(tag_img[0]['src'])
data = [title, author, score]
```

####  식단

```python
import requests

url = 'http://www.songok.es.kr/segio/meal_v2/meal_month.php'

month = [5, 6]
meal = []
for m in month:

    params = {'year': 2019, 'month':m}
    result = requests.get(url, params)

    soup = BeautifulSoup(result.content, 'lxml')

    tag_div = soup.find('div', class_='food_mon')
#     print(tag_div)
    tag_tr = tag_div.find_all('tr')
    for tr in tag_tr:
        tag_td = tr.find_all('td')
        for td in tag_td:
            tag_a = td.select('a')
            if len(tag_a) != 0:
                meal.append(tag_a[0].text)

new_meal = meal[0].replace('\n', ' ').split()

meal_list = []
for m in new_meal:
    if not '.' in m:
        if not '[' in m:
            meal_list.append(m)
print(meal_list)
```
