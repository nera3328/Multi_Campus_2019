```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
#가상의 분류 데이터 생성 메서드
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
X_train, y_train
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
```

달이라고 생각하자

parameter  
1) n_estimators : 트리의 수 in 숲 // 평가하는 estimateors 
2) criterion : gini or entropy  #gini계수는 높을수록 좋은거
3) max_depth : 최대 깊이 of tree.  
If None, until all leaves are pure or less than min_samples_split samples 안해주면 순수해질때까지/최소값까지  
4) min_samples_split  // 머신러닝 빠르게 해주려면   
5) max_features : 각 노드에서 선택하는 feature의 개수(무작위)  
6) n_jobs : 병렬 처리시 사용할 코어 수 (-1 : 모든 코어 사용)  // 머신러닝 빠르게 해주려면 병혈 사용.  




```python
model_forest = RandomForestClassifier(n_estimators=5, random_state=2)
model_forest.fit(X_train, y_train)
```

```python
model_forest.estimators_
```

```python
model_forest.classes_  
```

```python
model_forest.n_classes_  #2개다
```

```python
model_forest.n_features_  #2개다
```

```python
from matplotlib import pyplot as plt
import mglearn

#subplot은 6, figsize는 20,10
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
```

```python
axes
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), model_forest.estimators_)):
    
    ax.set_title("tree {}".format(i))
    mglearn.plots.plot_tree_partition(X, y, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(model_forest, X, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.show()
```
