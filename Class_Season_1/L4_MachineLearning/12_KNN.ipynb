제일 거리 가까운 애 3개 뽑아서 얘네는 어느클래스 비중이 높니?

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
data_wine= load_wine()
data_wine
```

data는 X, target_name은 y, class는 3개

```python
X, y = data_wine.data, data_wine.target
```

```python
import pandas as pd
X_df=pd.DataFrame(X,columns=data_wine.feature_names)
X_df.tail()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
X_train   #split 함
```

모델 학습 (K=3개)

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
print("Testest Accuracy: {:.2f}".format(model.score(X_test, y_test)))
```

예측하는 그림을 그리고파

```python
X_train = X_train[:, :2]
```

```python
from matplotlib import pyplot as plt
fig, axes = plt.subplots(1, 5, figsize=(20, 5)) #판을 5개로 쪼개주겠다. 사이즈는 figsize

k_neighbors = [i*2+1 for i in range(5)] # 1,3,5,7,9

for k, ax in zip(k_neighbors, axes):
    model_knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    mglearn.plots.plot_2d_separator(model_knn, X_train, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
    ax.set_title("{} neighbors".format(k))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc='best')   #loc: 범례의 위치, legend: 범례, axes: 자동으로 제일 가까운 애 뽑음
plt.show()
```
